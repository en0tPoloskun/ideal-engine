# 1. Постановка

Используя схему бегущего счёта и итерационные методы решить задачу:


\begin{equation}
\begin{cases}
\frac{\partial u}{\partial t} - \frac{1}{1+(u^2)} \cdot \frac{\partial u}{\partial x} = 0, \; -1\leq x\leq 0, \\
u(x, 0) = -x, \\
u(0, t) = 0.
\end{cases}
\end{equation}

#  2. Анализ 
## Построение характеристик

Получим уравнение характеристик для данной задачи:
\begin{equation}
\frac{dt}{1} = - ({1+(u^2)}) \cdot \frac{dx}{1}
\end{equation}

Отсюда переходим к интегрированию:

\begin{equation}
\int_{t_0}^t dt = - \int_{x_0}^x (1+(u^2))dx
\end{equation}

И, интегрируя, получаем
\begin{equation}
t - t_0 =  - (1+(u^2))(x - x_0)
\end{equation}

Подставим начальные и граничные условия в полученное равенство, тогда:

   1) $t_0$ = 0:  $t = - (1+x_0^2)(x-x_0)$
   
   2) $x_0$ = 0: $t = t_0 - x$
   
Имеем месейство кривых, по которым можем построить характеристики

```python
import numpy as np
import matplotlib.pyplot as plt

# характеристика при t0 = 0, x0 взята с определенным шагом на промежутке [-1; 0].
def char1(x):
    return [(-(1. + (x0 ** 2 )) * (x - x0)) for x0 in np.arange(-1., 0.1, 0.1)]

# характеристика при x0 = 0, t0 взята с определенным шагом на промежутке [0; 1].
def char2(x):
    return [(t0 - x) for t0 in np.arange(0, 1.1, 0.1)]

# Так как x принадлежит [-1; 0], создадим соответствующий массив
x_arr = np.arange(-1, 0.1, 0.1)

# И передадим массив в функции, отвечающие за характеристики
char1_arr = [char1(x) for x in x_arr]
char2_arr = [char2(x) for x in x_arr]

plt.subplot(1, 2, 1)
plt.ylim(-1,2.5)
plt.xlim(-1,0)
plt.plot(x_arr, char1_arr) 
plt.title('Характеристики при t_0 = 0')
plt.ylabel('t')
plt.xlabel('x')
plt.grid(True)
plt.subplot(1, 2, 2)
plt.ylim(-1,2.5)
plt.xlim(-1,0)
plt.plot(x_arr, char2_arr)
plt.title('Характеристики при x_0 = 0')
plt.xlabel('x')
plt.grid(True)
plt.show()
```
